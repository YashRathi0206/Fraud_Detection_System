TASK 1 — Data Loading & Exploratory Analysis

```python
#Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', 500)
```

```python
#Load Datasets
transaction = pd.read_csv("Data/train_transaction.csv")
identity = pd.read_csv("Data/train_identity.csv")
```

```python
#Merge Datasets
df = pd.merge(transaction, identity, on='TransactionID', how='left')

print("Merged Shape:", df.shape)
```

```python
# first 10 rows
df.head(10)
```

```python
#Class Imbalance analysis
print(df['isFraud'].value_counts())

sns.countplot(x='isFraud', data=df)
plt.title("Fraud vs Non-Fraud Transactions")
plt.savefig("Charts/Fraud_vs_Non-Fraud_Transactions.png", dpi=300, bbox_inches='tight')
plt.show()
```

## Observation
The dataset is highly imbalanced, with fraudulent transactions being much fewer than legitimate transactions.

```python
#Drop Columns >50% Missing
drop_cols = missing[missing > 50].index

df = df.drop(columns=drop_cols)

print("New Shape:", df.shape)
```

## Missing Value Strategy
Columns with more than 50% missing values were dropped to reduce noise and improve model performance.

TASK 2 — Preprocessing & Feature Engineering

```python
#Separate Numerical & Categorical
num_cols = df.select_dtypes(include=np.number).columns
cat_cols = df.select_dtypes(include='object').columns
```

```python
#Impute Missing Values
# Numerical → Median
for col in num_cols:
    df[col] = df[col].fillna(df[col].median())

# Categorical → Mode
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
#Label Encoding
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for col in cat_cols:
    df[col] = le.fit_transform(df[col].astype(str))
```

## Encoding Strategy
Label Encoding was used for categorical variables because many features contain high-cardinality categories. One-hot encoding would significantly increase dimensionality and memory usage.

```python
#Feature Engineering
# Feature 1: Amount to Mean Ratio
df['AmtToMeanRatio'] = df['TransactionAmt'] / df['TransactionAmt'].mean()

# Feature 2: Hour of Day
df['HourOfDay'] = (df['TransactionDT'] // 3600) % 24

# Feature 3: Transaction Amount Flag
df['HighAmountFlag'] = np.where(
    df['TransactionAmt'] > df['TransactionAmt'].median(),
    1,
    0
)

df[['AmtToMeanRatio', 'HourOfDay', 'HighAmountFlag']].head()
```

```python
#Train Test Split
from sklearn.model_selection import train_test_split

X = df.drop('isFraud', axis=1)
y = df['isFraud']

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,
    stratify=y,
    random_state=42
)
```

```python
#Class Ratio Before SMOTE
print(y_train.value_counts())
```

```python
#Apply SMOTE
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)

X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)
```

```python
#After SMOTE
print(y_train_smote.value_counts())
```

```python
#Scaling
from sklearn.preprocessing import RobustScaler

# Numerical columns excluding target
num_cols = X_train_smote.select_dtypes(include=np.number).columns

scaler = RobustScaler()

# Scale training data
X_train_smote[num_cols] = scaler.fit_transform(X_train_smote[num_cols])

# Scale test data
X_test[num_cols] = scaler.transform(X_test[num_cols])

print("Scaling completed successfully!")
```

TASK 3 — Model Training & Evaluation

```python
#Import ML Libraries
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score,
    confusion_matrix,
    classification_report,
    precision_recall_curve,
    auc,
    roc_curve
)

import lightgbm as lgb
from xgboost import XGBClassifier
from sklearn.ensemble import IsolationForest
```

```python
#Train LightGBM
lgb_model = lgb.LGBMClassifier(
    n_estimators=100,
    learning_rate=0.05,
    random_state=42
)

lgb_model.fit(X_train_smote, y_train_smote)
```

```python
#Train XGBoost
xgb_model = XGBClassifier(
    n_estimators=100,
    learning_rate=0.05,
    max_depth=6,
    random_state=42,
    eval_metric='logloss'
)

xgb_model.fit(X_train_smote, y_train_smote)
```

```python
#Train Isolation Forest
iso_model = IsolationForest(
    contamination=0.035,
    random_state=42
)

iso_model.fit(X_train_smote)
```

```python
#Predictions
# LightGBM
lgb_pred = lgb_model.predict(X_test)
lgb_prob = lgb_model.predict_proba(X_test)[:,1]

# XGBoost
xgb_pred = xgb_model.predict(X_test)
xgb_prob = xgb_model.predict_proba(X_test)[:,1]

# Isolation Forest
iso_pred = iso_model.predict(X_test)

# Convert IF output
iso_pred = np.where(iso_pred == -1, 1, 0)
```

```python
#Evaluation Function
def evaluate_model(y_true, y_pred, y_prob=None):

    print("Accuracy :", accuracy_score(y_true, y_pred))
    print("Precision:", precision_score(y_true, y_pred))
    print("Recall   :", recall_score(y_true, y_pred))
    print("F1 Score :", f1_score(y_true, y_pred))

    if y_prob is not None:
        print("ROC-AUC  :", roc_auc_score(y_true, y_prob))

        precision, recall, _ = precision_recall_curve(y_true, y_prob)
        pr_auc = auc(recall, precision)

        print("PR-AUC   :", pr_auc)
```

```python
#Evaluate Models
print("===== LIGHTGBM =====")
evaluate_model(y_test, lgb_pred, lgb_prob)

print("\n===== XGBOOST =====")
evaluate_model(y_test, xgb_pred, xgb_prob)

print("\n===== ISOLATION FOREST =====")
evaluate_model(y_test, iso_pred)
```

```python
#Confusion Matrix
fig, axes = plt.subplots(1,2, figsize=(12,5))

# LightGBM
sns.heatmap(
    confusion_matrix(y_test, lgb_pred),
    annot=True,
    fmt='d',
    ax=axes[0]
)

axes[0].set_title("LightGBM Confusion Matrix")

# XGBoost
sns.heatmap(
    confusion_matrix(y_test, xgb_pred),
    annot=True,
    fmt='d',
    ax=axes[1]
)

axes[1].set_title("XGBoost Confusion Matrix")
plt.savefig("Charts/LightGBM_vs_XGBoost_Confusion_Matrix.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
#ROC Curve
fpr_lgb, tpr_lgb, _ = roc_curve(y_test, lgb_prob)
fpr_xgb, tpr_xgb, _ = roc_curve(y_test, xgb_prob)

plt.figure(figsize=(8,6))

plt.plot(fpr_lgb, tpr_lgb, label='LightGBM')
plt.plot(fpr_xgb, tpr_xgb, label='XGBoost')

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.savefig("Charts/ROC_curve.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Precision-Recall values
precision_lgb, recall_lgb, _ = precision_recall_curve(y_test, lgb_prob)
precision_xgb, recall_xgb, _ = precision_recall_curve(y_test, xgb_prob)

# Plot
plt.figure(figsize=(8,6))

plt.plot(recall_lgb, precision_lgb, label='LightGBM')
plt.plot(recall_xgb, precision_xgb, label='XGBoost')

plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision Recall Curve")
plt.legend()

# Save figure
plt.savefig("Charts/precision_recall_curve.png", dpi=300, bbox_inches='tight')

# Show figure
plt.show()
```

## Model Comparison

LightGBM and XGBoost significantly outperform Isolation Forest in fraud detection performance. PR-AUC is especially important due to severe class imbalance.

TASK 4 — SHAP EXPLAINABLE AI

```python
#Import SHAP
import shap

shap.initjs()
```

```python
#SHAP Explainer
explainer = shap.TreeExplainer(lgb_model)

shap_values = explainer.shap_values(X_test)
```

```python
#SHAP Summary Plot
shap.summary_plot(
    shap_values,
    X_test,
    max_display=20,
    show=False
)
plt.savefig("Charts/shap_summary.png", dpi=300, bbox_inches='tight')

plt.show()
```

```python
#SHAP Feature Importance
shap.summary_plot(
    shap_values,
    X_test,
    plot_type="bar",
    max_display=20,
    show=False
)

plt.savefig("Charts/shap_Feature_Importance.png", dpi=300, bbox_inches='tight')

plt.show()
```

```python
#Fraud Transaction Explanation
fraud_index = y_test[y_test == 1].index[0]

shap.waterfall_plot(
    shap.Explanation(
        values=shap_values[fraud_index],
        base_values=explainer.expected_value,
        data=X_test.loc[fraud_index]
    ),
    show=False
)

plt.savefig("Charts/Fraud_Transaction_Explanation.png", dpi=300, bbox_inches='tight')

plt.show()
```

```python
# Find first legitimate transaction POSITION
normal_pos = np.where(y_test.values == 0)[0][0]

# SHAP waterfall plot
shap.waterfall_plot(
    shap.Explanation(
        values=shap_values[normal_pos],
        base_values=explainer.expected_value,
        data=X_test.iloc[normal_pos]
    ),
    show=False
)

# Save figure
plt.savefig(
    "charts/Legitimate_Transaction.png",
    dpi=300,
    bbox_inches='tight'
)

plt.show()
```

```python
#SHAP Dependence Plot
shap.dependence_plot(
    "TransactionAmt",
    shap_values,
    X_test,
    show= False
)

plt.savefig(
    "charts/Dependence_Plot.png",
    dpi=300,
    bbox_inches='tight'
)

plt.show()
```

## SHAP Insights

SHAP analysis shows that TransactionAmt, engineered ratio features, and time-based features contribute strongly to fraud predictions. High transaction amounts and unusual transaction timing increase fraud probability.

TASK 5 — Risk Segmentation & Fraud Pattern Analysis

```python
#Create Risk Scores
results_df = X_test.copy()

results_df['ActualFraud'] = y_test.values
results_df['FraudProbability'] = lgb_prob

results_df.head()
```

```python
#Create Risk Tiers
def assign_risk(prob):

    if prob >= 0.75:
        return "Critical Risk"

    elif prob >= 0.40:
        return "Suspicious"

    else:
        return "Clear"

results_df['RiskTier'] = results_df['FraudProbability'].apply(assign_risk)

results_df[['FraudProbability', 'RiskTier']].head()
```

```python
#Count Transactions per Tier
risk_counts = results_df['RiskTier'].value_counts()

print(risk_counts)
```

```python
#Risk Tier Bar Chart

plt.figure(figsize=(8,5))

sns.countplot(
    x='RiskTier',
    data=results_df
)

plt.title("Transactions by Risk Tier")

plt.savefig(
    "Charts/risk_tier_bar_chart.png",
    dpi=300,
    bbox_inches='tight'
)

plt.show()
```

```python
#Average Transaction Amount
avg_amt = results_df.groupby('RiskTier')['TransactionAmt'].mean()

print(avg_amt)
```

```python
#Fraud by Hour of Day
plt.figure(figsize=(10,5))

sns.countplot(
    x='HourOfDay',
    hue='RiskTier',
    data=results_df
)

plt.title("Fraud Risk by Hour of Day")

plt.savefig(
    "charts/hourly_risk_pattern.png",
    dpi=300,
    bbox_inches='tight'
)

plt.show()
```

```python
#Donut Chart
risk_counts = results_df['RiskTier'].value_counts()

plt.figure(figsize=(7,7))

plt.pie(
    risk_counts,
    labels=risk_counts.index,
    autopct='%1.1f%%'
)

centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()

fig.gca().add_artist(centre_circle)

plt.title("Risk Tier Distribution")

plt.savefig(
    "charts/risk_tier_donut_chart.png",
    dpi=300,
    bbox_inches='tight'
)

plt.show()
```

```python
#Fraud Patterns
critical = results_df[
    results_df['RiskTier'] == 'Critical Risk'
]

print("Average Amount:")
print(critical['TransactionAmt'].mean())

print("\nAverage Hour:")
print(critical['HourOfDay'].mean())

print("\nFraud Ratio:")
print(critical['ActualFraud'].mean())
```

## Fraud Pattern Analysis

Critical Risk transactions tend to:
- Have higher transaction amounts
- Occur during unusual hours
- Show significantly higher fraud probability
